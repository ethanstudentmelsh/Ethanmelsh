```python
# plot DataFrame columns
d = {'col1':np.random.randint(100,size=20),
     'col2': np.random.randint(100,size=20)}
d
```

```python
df = DataFrame(d)
df
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(df['col1'], 'k.', label='col1')
ax.plot(df['col2'], 'ko--', label='col2')
ax.legend(loc='best')
```

```python
# Ticks, Labels, and Legends
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
```

```python
data = randn(1000).cumsum()
data
```

```python
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['One','Two','Three', 'Four','Five'], rotation=30, fontsize='small')

ax.set_title("My First Matplotlip Plot With Ticks")
ax.set_xlabel('Stages')
ax.set_ylabel('values')
fig
```

```python
props = ( 'title': "My First Matplotlip Plot With Ticks"
           'xlabel' : "Stages"
           'ylabel' : "Values"
        
        )

ax.set(**prop)
fig
```
